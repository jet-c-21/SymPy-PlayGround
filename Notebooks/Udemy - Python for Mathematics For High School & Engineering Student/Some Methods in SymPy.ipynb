### Identity Matrix

```python
def gen_identy_arr(i, j):
    if i == j:
        return 1
    else:
        return 0
```

```python
I3 = sp.Matrix(3, 3, gen_identy_arr)
I3
```

```python
I3.det()
```

```python
m.eigenvects()
```

```python
m.eigenvals()
```

```python
arr_d = [
    [1,0,0],
    [0,2,0],
    [0,0,3],
]
D = sp.Matrix(arr_d)
D
```

```python
D.eigenvals()
```

```python
arr_e = [
    [1,0,0],
    [0,1,0],
    [0,0,3],
]
E = sp.Matrix(arr_e)
E
```

```python
E.eigenvals()
```

## Solve Linear Equation

```python
x, y = sp.symbols('x y')
```

```python
eq1 = sp.Eq(2*x+y, 5)
eq1
```

```python
eq2 = sp.Eq(x+4*y, 6)
eq2
```

```python
sp.solve((eq1, eq2))
```
